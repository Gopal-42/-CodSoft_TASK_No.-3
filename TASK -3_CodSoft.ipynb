*Importing Important Libraries*

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

*Importing DATASET*

```python
data_cc = pd.read_csv("/content/creditcard.csv")
data_cc.head()
```

```python
data_cc.tail()
```

```python
data_cc.info()
```

```python
data_cc.isnull().sum()
```

**Distribution of transactions**

```python
data_cc['Class'].value_counts()
```

*DATASET Analysing*



0=> Normal Transaction                      
1=> Fraudulent Transaction

```python
legit = data_cc[data_cc.Class == 0]
fraud = data_cc[data_cc.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values of both transactions
data_cc.groupby('Class').mean()
```

**Sampling**

Number of Fraudulent Transaction  ==> 198

```python
legit_sam= legit.sample(n=198)
```

```python
new_data = pd.concat([legit_sam, fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

*Modelling and Training*

```python
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

*Model Prediction*

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Training Data Acuracy:', train_data_accuracy)
```

```python
x_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(x_test_prediction, Y_test)
```

```python
print('Test Data Acuracy:', test_data_accuracy)
```
